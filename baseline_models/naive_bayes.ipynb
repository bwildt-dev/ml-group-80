# Naive Bayes Classifier - Likelihood to Smoke
This notebook serves as an indicator for the baseline performance of the dataset currently used.

## Imports

```python
from sklearn.naive_bayes import GaussianNB
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    confusion_matrix,
    roc_curve,
    RocCurveDisplay
)
```

## Initiation

```python
model = GaussianNB()

#Importing training set
X_train = pd.read_csv("../data/trainX.csv")
y_train = pd.read_csv("../data/trainY.csv")

#Importing validation set
X_valid = pd.read_csv("../data/validX.csv")
y_valid = pd.read_csv("../data/validY.csv")
```

```python
X_train.head()
```

## Model Training

```python
model.fit(X_train, np.ravel(y_train, order='C'))
```

## Evaluation

```python
y_pred = model.predict(X_valid)
acc = accuracy_score(y_pred, y_valid)
f1 = f1_score(y_pred, y_valid, average='binary')

print("Accuracy:", acc)
print("F1 Score:", f1)
```

### Confusion Matrix

```python
cm = confusion_matrix(y_valid, y_pred, labels=model.classes_)
matrix_plot = sns.heatmap(cm, annot=True, fmt="d", cmap="ocean")

matrix_plot.set_title("Confusion Matrix for Naive Bayes Classifier")
matrix_plot.set_xlabel("Predicted Smoker", fontsize=14, labelpad=20)
matrix_plot.xaxis.set_ticklabels(['Negative', 'Positive'])

matrix_plot.set_ylabel("Actual Smoker", fontsize=14, labelpad=20)
matrix_plot.yaxis.set_ticklabels(['Negative', 'Positive'])

plt.show()
```

### ROC-curve

```python
y_proba = model.predict_proba(X_valid)

fpr, tpr, _ = roc_curve(y_valid, y_proba[:, 1])
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
roc_display.figure_.set_size_inches(5,5)
plt.plot([0, 1], [0, 1], color = 'g')
```
